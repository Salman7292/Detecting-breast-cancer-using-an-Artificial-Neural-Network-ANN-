# **Detecting breast cancer using an Artificial Neural Network (ANN)**

## **Know About Our DataSet**

1. **id**: Unique identifier for each patient.
2. **diagnosis**: **Target Column**. Indicates whether the tumor is benign (B) or malignant (M).

The following columns are computed from the digitized image of a fine needle aspirate (FNA) of a breast mass. Each feature is calculated for three different aspects: mean value, standard error (SE), and "worst" or largest value (mean of the three largest values). The columns are grouped accordingly:

### Mean Values of Features
3. **radius_mean**: Mean of distances from the center to points on the perimeter.
4. **texture_mean**: Mean of gray-scale values.
5. **perimeter_mean**: Mean size of the core tumor.
6. **area_mean**: Mean area of the tumor.
7. **smoothness_mean**: Mean of local variation in radius lengths.
8. **compactness_mean**: 
9. **concavity_mean**: Mean of the severity of concave portions of the contour.
10. **concave points_mean**: Mean of the number of concave portions of the contour.
11. **symmetry_mean**: Mean symmetry of the tumor.
12. **fractal_dimension_mean**: Mean of "coastline approximation" 

### Standard Error of Features
13. **radius_se**: Standard error of the distances from the center to points on the perimeter.
14. **texture_se**: Standard error of gray-scale values.
15. **perimeter_se**: Standard error of the size of the core tumor.
16. **area_se**: Standard error of the area of the tumor.
17. **smoothness_se**: Standard error of local variation in radius lengths.
18. **compactness_se**: Standard error of 
19. **concavity_se**: Standard error of the severity of concave portions of the contour.
20. **concave points_se**: Standard error of the number of concave portions of the contour.
21. **symmetry_se**: Standard error of symmetry of the tumor.
22. **fractal_dimension_se**: Standard error of "coastline approximation" 

### "Worst" or Largest Values of Features
23. **radius_worst**: Largest mean of distances from the center to points on the perimeter.
24. **texture_worst**: Largest mean of gray-scale values.
25. **perimeter_worst**: Largest mean size of the core tumor.
26. **area_worst**: Largest mean area of the tumor.
27. **smoothness_worst**: Largest mean of local variation in radius lengths.
28. **compactness_worst**: Largest mean of 
29. **concavity_worst**: Largest mean of the severity of concave portions of the contour.
30. **concave points_worst**: Largest mean of the number of concave portions of the contour.
31. **symmetry_worst**: Largest mean symmetry of the tumor.
32. **fractal_dimension_worst**: Largest mean of "coastline approximation"

# **Now Defining the Hyperparamter for the Atrifical Neural Network**

```python
# Hyperparameters for the neural network
input_size = 30  # Number of input features
hidden_layer1 = 60  # Number of neurons in the first hidden layer
hidden_layer2 = 30  # Number of neurons in the second hidden layer
output_layer = 1   # Number of output neurons (for binary classification)
batch_size = 14   # Batch size for training
```

# **Now Buliding the Model**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Define early stopping
early_stopping = EarlyStopping(
    monitor='val_loss',
    min_delta=0.1,
    patience=4,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False
)
```

```python
def get_model():

  model = Sequential()  # Create a sequential model

  # Add the first hidden layer with ReLU activation
  model.add(Dense(hidden_layer1, input_dim=input_size, activation='relu')) 


  # Add the output layer with sigmoid activation for binary classification
  model.add(Dense(output_layer, activation='sigmoid'))

  # Print model summary for debugging
  model.summary() 

  # Compile the model with Adam optimizer, binary crossentropy loss, and accuracy metric
  model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
  return model
```

```python
Model=get_model()
```

```python
from time import time
```

# **Now Train the Model**

```python
# Record the start time
start_time = time()

# Train the model
History=Model.fit(
    X_train_Scaled,  # Assuming X_train_scaled is the correctly named variable
    y_train,
    batch_size=batch_size,
    epochs=50,        # Corrected argument from 'epoch' to 'epochs'
    validation_split=0.20,  # Corrected validation split argument
    verbose=1,         # Corrected 'verbse' to 'verbose'
    callbacks=[early_stopping]  # Include early stopping callback
)

# Record the end time
end_time = time()

# Calculate the total training time
total_time = end_time - start_time
total_time = round(total_time, 1)

# Print the total training time
print(f"Total Training Time: {total_time} seconds")

```
